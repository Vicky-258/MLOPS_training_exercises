```python
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()

wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_df['target'] = wine.target

print("Features of the wine dataset:")
print(wine.feature_names)
print("Target names (classes):")
print(wine.target_names)
print("\nFirst 5 rows of the wine dataset:")
display(wine_df.head())
```

```python
from sklearn.model_selection import train_test_split

X = wine.data
y = wine.target

X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set size: {X_train_wine.shape[0]} samples")
print(f"Test set size: {X_test_wine.shape[0]} samples")
```

```python
from sklearn.naive_bayes import GaussianNB

wine_classifier = GaussianNB()

wine_classifier.fit(X_train_wine, y_train_wine)
```

```python
from sklearn.metrics import accuracy_score, classification_report

wine_predictions = wine_classifier.predict(X_test_wine)

wine_accuracy = accuracy_score(y_test_wine, wine_predictions)

print(f"Model Accuracy on Wine Test Set: {wine_accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test_wine, wine_predictions, target_names=wine.target_names))
```
