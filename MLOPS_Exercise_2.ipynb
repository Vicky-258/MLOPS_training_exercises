```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('hiring.csv')

print("DataFrame Info:")
df.info()
print("\nFirst 5 rows of the DataFrame:")
display(df.head())
```

```python
experience_mapping = {
    'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5,
    'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10,
    'eleven': 11, 'twelve': 12
}
df['experience'] = df['experience'].replace(experience_mapping)

df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

median_experience = df['experience'].median()
df['experience'].fillna(median_experience, inplace=True)

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

median_test_score = X['test_score(out of 10)'].median()
X['test_score(out of 10)'].fillna(median_test_score, inplace=True)


model = LinearRegression()
model.fit(X, y)

print("Model training complete.")
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
new_data = pd.DataFrame({
    'experience': [2, 12],
    'test_score': [9, 10],
    'interview_score': [6, 10]
})

predictions = model.predict(new_data)

for i, (exp, test, interview, pred) in enumerate(zip(new_data['experience'], new_data['test_score'], new_data['interview_score'], predictions)):
    print(f"Prediction {i+1}:\n  Experience: {exp} yr, Test Score: {test}, Interview Score: {interview} -> Predicted Salary: {pred:.2f}")
```
